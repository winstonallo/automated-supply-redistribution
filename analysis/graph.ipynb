```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv('outputafterdemand.csv', sep=';')
df2 = pd.read_csv('outputbefore.csv', sep=';')
result = df1.merge(df2, on=['Store','Item'], how='left')
result['Before'] = result['Quantity_y']
result['After'] = result['Quantity_x']
result = result.drop(['Quantity_x', 'Quantity_y'], axis=1)
```

```python

sns.kdeplot(result['Before'], label="Before", bw_method=0.2, color='b',fill=True, alpha=0.5)
sns.kdeplot(result['After'], label="After" , bw_method=0.32, color='#EC008B',fill=True, alpha=0.5)

# Customize the plot
plt.xlabel("Deviation from demand")
plt.xticks(np.arange(-10, 11, 2))
plt.ylabel("Density")
plt.title("Density Plot of Deviation from Demand Before and After our Model", fontsize=10)
plt.xlim(-10, 10)
plt.axvline(0, color='black', linestyle='--', alpha=0.5)
plt.legend(loc = 'upper left')
# improve resolution
plt.savefig('output.png', dpi=1000)
plt.show()
```
